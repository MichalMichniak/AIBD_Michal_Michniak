```python
import pandas as pd
df = pd.read_csv("../OrginalData/drinks.csv")
```

```python
col_names = ["country","beer","spirit","wine","pure_alcohol"]
df.columns = col_names
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg as sp
plt.plot(df["pure_alcohol"],df["beer"], ".b")
plt.title("Comparation of beer consumption to\ntotal alcochol consumption")
plt.ylabel("cans of beer")
plt.xlabel("litres of pure alcochol")
A = np.array([df["pure_alcohol"],np.ones(len(df["pure_alcohol"]))])
Q,R = np.linalg.qr(A.T)
a,b = sp.solve_triangular(R,Q.T@np.array(df["beer"]))
plt.plot([0,450],[b,450*a + b],"r", label = f"least square regresion y = {a:.1f}x + ({b:.1f})")
plt.xlim([0,15])
plt.ylim([0,400])
plt.legend()
plt.show()
```

```python
plt.plot(df["pure_alcohol"],df["wine"], ".b")
plt.title("Comparation of wine consumption to\ntotal alcochol consumption")
plt.ylabel("glass of wine")
plt.xlabel("litres of pure alcochol")
A = np.array([df["pure_alcohol"],np.ones(len(df["pure_alcohol"]))])
Q,R = np.linalg.qr(A.T)
a,b = sp.solve_triangular(R,Q.T@np.array(df["wine"]))
plt.plot([0,450],[b,450*a + b],"r", label = f"least square regresion y = {a:.1f}x + ({b:.1f})")
plt.xlim([0,15])
plt.ylim([0,400])
plt.legend()
plt.show()
```

```python
plt.plot(df["pure_alcohol"],df["spirit"], ".b")
plt.title("Comparation of spirits consumption to\ntotal alcochol consumption")
plt.ylabel("measure of spirit")
plt.xlabel("litres of pure alcochol")
A = np.array([df["pure_alcohol"],np.ones(len(df["pure_alcohol"]))])
Q,R = np.linalg.qr(A.T)
a,b = sp.solve_triangular(R,Q.T@np.array(df["spirit"]))
plt.plot([0,450],[b,450*a + b],"r", label = f"least square regresion y = {a:.1f}x + ({b:.1f})")
plt.xlim([0,15])
plt.ylim([0,438])
plt.legend()
plt.show()
```
