```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
import Command_Files.main as m
m.clean()
df = pd.read_csv("../Analysis Data/wielkopolskie.csv",index_col=0)
df2 = pd.read_csv("../Orginal Data/14_WIELKOPOLSKIE.csv",index_col=0)
# imputacja ze względu na wiek
df2 = df2[~df2["Wiek kupującego"].isnull()]
df.head()
```

```python
marki = df["Marka"].unique()
plt.bar(marki,np.array([df[df["Marka"] == i]["Marka"].count() for i in marki]))
plt.title("Udział firm w ankietach")
plt.ylabel("liczba ocenionych sztuk")
plt.show()
```

```python
plt.hist(df2["Wiek kupującego"],bins = len(df2["Wiek kupującego"].unique()))
plt.title("Wiek ankietowanych")
plt.ylabel("liczba ocen")
plt.xlabel("Wiek w latach")
plt.show()
```

```python
k_df2 = df2[df2["Płeć kupującego"] == "K"]
plt.hist(k_df2["Wiek kupującego"],bins = len(k_df2["Wiek kupującego"].unique()))
plt.title("Wiek ankietowanych kobiet")
plt.ylabel("liczba ocen")
plt.xlabel("Wiek w latach")
plt.show()
```

```python
m_df2 = df2[df2["Płeć kupującego"] == "M"]
plt.hist(m_df2["Wiek kupującego"],bins = len(m_df2["Wiek kupującego"].unique()))
plt.title("Wiek ankietowanych mężczyzn")
plt.ylabel("liczba ocen")
plt.xlabel("Wiek w latach")
plt.show()
```

```python
plt.bar(["kobiety", "mężczyźni"],np.array([k_df2["Płeć kupującego"].count(),m_df2["Płeć kupującego"].count()]))
plt.title("ilość wystawionych ocen ze względu na płeć")
plt.ylabel("liczba ocenionych sztuk")
plt.show()
```

```python
oceny = list(df.Ocena.unique())
oceny.sort()
plt.bar(oceny,np.array([df[df.Ocena == i].Ocena.count() for i in oceny]), width = 0.45)
plt.title("rozkład wystawionych ocen")
plt.ylabel("liczba ocenionych sztuk")
plt.xlabel("oceny")
plt.show()
```

```python
oceny = list(df.Ocena.unique())
oceny.sort()
for name in df.Marka.unique():
    plt.bar(oceny,np.array([df[(df.Ocena == i) & (df.Marka == name)].Ocena.count() for i in oceny]), width = 0.45)
    plt.title(f"rozkład wystawionych ocen dla firmy {name}")
    plt.ylabel("liczba ocenionych sztuk")
    plt.xlabel("oceny")
    plt.show()
```

```python
oceny = list(df.Ocena.unique())
oceny.sort()
for name in [("K","kobiety"), ("M", "mężczyzn")]:
    plt.bar(oceny,np.array([df[(df.Ocena == i) & (df["Płeć kupującego"] == name[0])].Ocena.count() for i in oceny]), width = 0.45)
    plt.title(f"rozkład wystawionych ocen przez {name[1]}")
    plt.ylabel("liczba ocenionych sztuk")
    plt.xlabel("oceny")
    plt.show()
```

```python
dni = df["Dni od zakupu"].unique()
plt.hist(df["Dni od zakupu"],bins = len(df["Dni od zakupu"].unique()))
plt.title("rozkład liczby dni po których\nzostała wystawiona ocena")
plt.ylabel("liczba ocen")
plt.xlabel("dni")
plt.show()
```
